```python
from rdkit import Chem
from rdkit.Chem import rdChemReactions as Reactions
from rdkit.Chem import Draw
import json

with open('./out_beam10_best10.txt', 'r') as f:
    lines = f.readlines()
with open('../data/label_template.json', 'r') as f:
    d = json.load(f)
```

```python
template = None
for line in lines:
    line = line.split(' ')
    t = str(line[0].strip('[]'))
    if template != t:
        template = t
        template_rxn = Reactions.ReactionFromSmarts(d[template])
        print('='*100)
        display(template_rxn)
    rsmi = ''.join(line[1:]).replace(' ', '')
    try:
        rxn = Reactions.ReactionFromSmarts(rsmi, useSmiles=True)
        display(rxn)
    except:
        pass
```
